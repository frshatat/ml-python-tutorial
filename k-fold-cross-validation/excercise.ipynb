```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
import numpy as np
from sklearn.datasets import load_iris

iris = load_iris()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.3)
```

```python
def get_score(model, X_train, X_test, y_train, y_test):
    model.fit(X_train, y_train)
    return model.score(X_test, y_test)
```

```python
lr_score = get_score(LogisticRegression(), X_train, X_test, y_train, y_test)
lr_score
```

```python
from sklearn.model_selection import StratifiedKFold
kf = StratifiedKFold(n_splits=3, shuffle=True, random_state=1)
```

```python
svm_score = get_score(SVC(), X_train, X_test, y_train, y_test)
svm_score
```

```python
rf_score = get_score(RandomForestClassifier(n_estimators=200), X_train, X_test, y_train, y_test)
rf_score
```

```python
scores_lr = []
scores_svm = []
scores_rf = []
for train_index, test_index in kf.split(iris.data, iris.target):
    X_train, X_test = iris.data[train_index], iris.data[test_index]
    y_train, y_test = iris.target[train_index], iris.target[test_index]
    scores_lr.append(get_score(LogisticRegression(), X_train, X_test, y_train, y_test))
    scores_svm.append(get_score(SVC(), X_train, X_test, y_train, y_test))
    scores_rf.append(get_score(RandomForestClassifier(n_estimators=200), X_train, X_test, y_train, y_test))
```

```python
scores_lr
```

```python
scores_svm
```

```python
scores_rf
```

```python
from sklearn.model_selection import cross_val_score
```

```python
scores_lr = cross_val_score(LogisticRegression(max_iter=1000), iris.data, iris.target, cv=3)
np.average(scores_lr)
```

```python
scores_svm = cross_val_score(SVC(), iris.data, iris.target, cv=3)
np.average(scores_svm)
```

```python
scores_rf = cross_val_score(RandomForestClassifier(n_estimators=200), iris.data, iris.target, cv=3)
np.average(scores_rf)
```
