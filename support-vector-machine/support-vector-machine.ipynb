```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
dir(iris)
```

```python
iris.feature_names
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df.head()
```

```python
df['target'] = iris.target
df.head()
```

```python
iris.target_names
```

```python
df[df.target == 2].head()
```

```python
df['flower_name'] = df.target.apply(lambda x: iris.target_names[x])
df.head()
```

```python
from matplotlib import pyplot as plt
```

```python
%matplotlib inline
```

```python
df0 = df[df.target == 0]
df1 = df[df.target == 1]
df2 = df[df.target == 2]    
```

```python
plt.scatter(df0['sepal length (cm)'], df0['sepal width (cm)'], color='blue', label=iris.target_names[0])
plt.scatter(df1['sepal length (cm)'], df1['sepal width (cm)'], color='red', label=iris.target_names[1])
plt.scatter(df2['sepal length (cm)'], df2['sepal width (cm)'], color='green', label=iris.target_names[2])
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.legend()
```

```python
plt.scatter(df0['petal length (cm)'], df0['petal width (cm)'], color='blue', label=iris.target_names[0])
plt.scatter(df1['petal length (cm)'], df1['petal width (cm)'], color='red', label=iris.target_names[1])
plt.scatter(df2['petal length (cm)'], df2['petal width (cm)'], color='green', label=iris.target_names[2])
plt.xlabel('petal length (cm)')
plt.ylabel('petal width (cm)')
plt.legend()
```

```python
x = df.drop(['target', 'flower_name'], axis='columns')
```

```python
y = df.target
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.svm import SVC
model = SVC()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
